```
train_data = sqlContext.read.format('com.databricks.spark.csv').options(header='true', inferSchema='true').load('/FileStore/tables/train_set-51e11.csv')

test_data = sqlContext.read.format('com.databricks.spark.csv').options(header='true', inferSchema='true').load('/FileStore/tables/test_set-b5f57.csv')

display(train_data)
```

This first description of the dataset gives us several informations :
- we have 10 quantitative variables, uncentered and not normalized
- we have boolean variables regarding the soil type and the presence of wilderness area.

```
from pyspark.sql import functions as F

display(train_data.describe())
```

This second insight confirms that some of the quantative variables have a very high variance (horizontal distance to roadways and horizontal distance to firepoints mainly)

We also note that some variables have a very low standard deviation : I chose to drop variables with a standard deviation lower to 0.02, because they give very poor information.

```
train_data.drop(train_data.Soil_Type15)
train_data.drop(train_data.Soil_Type8)
train_data.drop(train_data.Soil_Type7)
```

For the variables with a very high standard deviation, I normalized them, dividing by the standard deviation.

```
std_HdtR = (training_data.select(F.stddev("Horizontal_Distance_To_Roadways")).collect())[0][0]
std_HdtFP = (training_data.select(F.stddev("Horizontal_Distance_To_Fire_Points")).collect())[0][0]
```

Moreovere, one can observe that the hillshade variables, according to data description in the Kaggle are supposed to be between 0 and 255. I tried to filter them because some occurences are negative and have no meaning.

```
train_data = train_data.filter(train_data.Hillshade_9am >= 0)
train_data = train_data.filter(train_data.Hillshade_3pm >= 0)
train_data = train_data.filter(train_data.Hillshade_Noon >= 0)
```

The following section is dedicated to the creation of new explicative variables.

```
train_data = train_data.withColumn('Abs_H_D_t_H', train_data.Horizontal_Distance_To_Hydrology ** 2 ** 0.5)
train_data = train_data.withColumn('Abs_V_D_t_H', train_data.Vertical_Distance_To_Hydrology ** 2 ** 0.5)
train_data = train_data.withColumn('Abs_Az', train_data.Aspect ** 2** 0.5)
train_data = train_data.withColumn('mean_shade', (train_data.Hillshade_9am + train_data.Hillshade_3pm  + train_data.Hillshade_Noon  ) /3)
train_data = train_data.withColumn('Abs_H_D_t_R', train_data.Horizontal_Distance_To_Roadways ** 2** 0.5)
train_data = train_data.withColumn('Abs_H_D_t_FP', train_data.Horizontal_Distance_To_Fire_Points ** 2** 0.5)
train_data = train_data.withColumn('SlopeSquared', train_data.Slope ** 2** 0.5)
train_data = train_data.withColumn('test', train_data.Horizontal_Distance_To_Fire_Points - train_data.Horizontal_Distance_To_Hydrology)
train_data = train_data.withColumn('hydrodiff', train_data.Vertical_Distance_To_Hydrology - train_data.Horizontal_Distance_To_Hydrology)
train_data = train_data.withColumn('elevation_sq', train_data.Elevation ** 2)
train_data = train_data.withColumn('elevation_log', F.log(train_data.Elevation))
train_data = train_data.withColumn('std_rail', train_data.Horizontal_Distance_To_Roadways / std_HdtR)
train_data = train_data.withColumn('std_dist', train_data.Horizontal_Distance_To_Fire_Points / std_HdtFP)
```

```
test_data = test_data.withColumn('Abs_H_D_t_H', test_data.Horizontal_Distance_To_Hydrology ** 2 ** 0.5)
test_data = test_data.withColumn('Abs_V_D_t_H', test_data.Vertical_Distance_To_Hydrology ** 2 ** 0.5)
test_data = test_data.withColumn('Abs_Az', test_data.Aspect ** 2 ** 0.5)
test_data = test_data.withColumn('mean_shade', (test_data.Hillshade_9am + test_data.Hillshade_3pm  + test_data.Hillshade_Noon  ) /3)
test_data = test_data.withColumn('Abs_H_D_t_R', test_data.Horizontal_Distance_To_Roadways ** 2 ** 0.5)
test_data = test_data.withColumn('Abs_H_D_t_FP', test_data.Horizontal_Distance_To_Fire_Points ** 2 ** 0.5)
test_data = test_data.withColumn('SlopeSquared', test_data.Slope ** 2 ** 0.5)
test_data = test_data.withColumn('test', test_data.Horizontal_Distance_To_Fire_Points - test_data.Horizontal_Distance_To_Hydrology)
test_data = test_data.withColumn('hydrodiff', test_data.Vertical_Distance_To_Hydrology - test_data.Horizontal_Distance_To_Hydrology)
test_data = test_data.withColumn('elevation_sq', test_data.Elevation ** 2)
test_data = test_data.withColumn('elevation_log', F.log(test_data.Elevation))
test_data = test_data.withColumn('std_rail', test_data.Horizontal_Distance_To_Roadways / std_HdtR)
test_data = test_data.withColumn('std_dist', test_data.Horizontal_Distance_To_Fire_Points / std_HdtFP)
```

```
print('Train data size: {} rows, {} columns'.format(train_data.count(), len(train_data.columns)))
print('Test data size: {} rows, {} columns'.format(test_data.count(), len(test_data.columns)))
```

I chose to assemble multiple vectorAssemblers, to be able to test different models.

```
from pyspark.ml import Pipeline
```

```
# Evaluator definition : for testing models before submitting
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(predictionCol="prediction", labelCol = "Cover_Type")
```

**Model 1 : decision tree with existing features **

Only the variables with very low variance are deleted.

```
from pyspark.ml.classification import DecisionTreeClassifier
```

```
# Default parameters
dt1 = DecisionTreeClassifier( labelCol="Cover_Type")

# Chain vecAssembler and classificaiton model 
pipeline = Pipeline(stages=[vector_assembler, dt1])

# Run stages in pipeline with the train data
model = pipeline.fit(training_data)
```

```
predictions_training1 = model.transform(validation_data)
```

```
evaluator.evaluate(predictions_training1)
```

Once we have the results on our training set, we can reeveluate the model on the whole training set to export results on the real test set.

```
# Reevaluate model
model = pipeline.fit(train_data)

# Make predictions on testData
predictions = model.transform(test_data)

# Cast predictions to 'int' to mach the data type expected by Kaggle
predictions = predictions.withColumn("Cover_Type", predictions["prediction"].cast("int"))
```

This classifier gives very poor results. 

Then, I tried to cross validate on some parameters of the classifier. Due to databricks limitations, I was only able to cross validate on 
- maxDepth = [10, 20, 30] since 30 is a limitation in pyspark
- impurity measure gini or entropy

** Cross validation **

```
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
paramGrid = ParamGridBuilder().addGrid(dt1.maxDepth, [10,20,30]).addGrid(dt1.impurity, ['gini', 'entropy']).build()

crossval = CrossValidator(estimator=pipeline,
                        estimatorParamMaps=paramGrid,
                        evaluator=evaluator,
                        numFolds=3)
```

```
cvModel = crossval.fit(training_data)
```

```
predictions_cv = cvModel.transform(validation_data)
evaluator.evaluate(predictions_cv)
```

This model is obviously best fitted than the default decision tree. Due to the extremely high time needed to run a simple cross validation, I will therefore consider a maxDepth = 30 and an impurity measure which is entropy.

```
cvModel_submit = crossval.fit(train_data)
predictions = cvModel_submit.transform(test_data)
predictions = predictions.withColumn("Cover_Type", predictions["prediction"].cast("int"))
```

```
bestModelCv = cvModel.bestModel
```

```
print(bestModelCv.stages[1].extractParamMap())
```

** Next step is a decision tree with all created features **

```
dt2 = DecisionTreeClassifier( maxDepth=30, maxBins=250, labelCol="Cover_Type", impurity = 'entropy')

pipeline_dt2 = Pipeline(stages=[vector_assembler4, dt2])
```

```
model_dt2 = pipeline_dt2.fit(training_data)
```

```
predictions_dt2 = model_dt2.transform(validation_data)
evaluator.evaluate(predictions_dt2)
```

```
# Train again on whole set
model_dt2_submit = pipeline_dt2.fit(train_data)
predictions = model_dt2_submit.transform(test_data)
```

It can be useful to take a look at the feature importance in the model. In every tree I trained it appeared that the most important variable is Elevation. Horizontal distance seems to have a high importance as well. Boolean variables have a low level, but maybe this is because each one of them is binary.

```
model_dt2_submit.stages[1].featureImportances
```
